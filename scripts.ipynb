```python
#Find a String

def count_substring(string, sub_string):
    lenght_sub=len(sub_string)
    a=0
    counter=0
    for i in range(len(string)-lenght_sub+1):
        for s in range(lenght_sub):
            if string[i+s]==sub_string[s]:
                a += 1
                if a==lenght_sub:
                    counter += 1
                    a = 0
        a=0
    return counter
```

```python
#Text Wrap

def wrap(string, max_width):
    string_2=''
    for i in range(0,len(string),max_width):
        string_2 += string[i:(i+max_width)] + '\n'
    
    return string_2
```

```python
#Capitalize!

def solve(s):
    lst=[]
    i=0
    while i < len(s):
        if s[i]==' ':
            lst.append(s[i])
            i += 1
        else:
            start=i
            while i<len(s) and s[i] != ' ':
                i += 1
            b=s[start:i]
            new_word =b[0].upper() + b[1:].lower()
            lst.append(new_word)
    
    j=''.join(lst)
    return j 
```

```python
#The Minion Game

def minion_game(string):
    score_S=0
    score_K=0
    i=0
    while i < len(string):
        if string[i] in 'BCDFGHJKLMNPQRSTVWXYZ':
            score_S += (len(string)-i)
            i += 1
        elif string[i] in 'AEIOU':
            score_K += (len(string)-i)
            i += 1
        
    
    if score_K<score_S:
        print('Stuart {}'.format(score_S))
    elif score_K>score_S:
        print('Kevin {}'.format(score_K))
    else:
        print('Draw')
```

```python
#Piling Up!

from collections import deque

T=int(input())

for i in range(T):
    n=int(input())
    lst=list(map(int,input().split()))
    d=deque(lst)
    o=max(d[0],d[-1])
    
   
    while d:
        if len(d) == 1:
            if d[0] <= o:
                print('Yes')
            else:
                print('No')
            break

        current_max = max(d[0], d[-1])

        if current_max <= o:
            o = current_max
            if d[0] == current_max:
                d.popleft()
            else:
                d.pop()
        else:
            print('No')
            break

    else:
        print('Yes')
```

```python
#Validating UID

import re

t=int(input())

p=r'^(?=(?:.*[A-Z]){2,})(?=(?:.*\d){3,})[A-Za-z0-9]{10}$'

for i in range(t):
    s=input()
    
    if re.match(p,s) and len(set(s))==10:
        print('Valid')
    else:
        print('Invalid')
```

```python
#Validating Credit Card Numbers

import re

p= r'^(?!.*(\d)(?:-?\1){3})[4,5,6]\d{15}$|^(?!.*(\d)(?:-?\2){3})(\d{4}-){3}\d{4}$'


n=int(input())

for i in range(n):
    s=input()
    if re.match(p,s):
        print('Valid')
    else:
        print('Invalid')
```

```python
#Number Line Jumps

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'kangaroo' function below.
#
# The function is expected to return a STRING.
# The function accepts following parameters:
#  1. INTEGER x1
#  2. INTEGER v1
#  3. INTEGER x2
#  4. INTEGER v2
#

def kangaroo(x1, v1, x2, v2):
    # Write your code here
    for i in range(1,5000):
        if x1+v1*i==x2+v2*i:
            return 'YES'
        
    return 'NO'

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    first_multiple_input = input().rstrip().split()

    x1 = int(first_multiple_input[0])

    v1 = int(first_multiple_input[1])

    x2 = int(first_multiple_input[2])

    v2 = int(first_multiple_input[3])

    result = kangaroo(x1, v1, x2, v2)

    fptr.write(result + '\n')

    fptr.close()
```

```python
#Recursive Digit Sum

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'superDigit' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. STRING n
#  2. INTEGER k
#

def superDigit(n, k):
    
    n=sum(list(map(int,n)))*k
    
    def obt_sd(n):
        if n<10:
            return n
        else:
            return obt_sd(sum(list(map(int,str(n)))))
            
    return obt_sd(n)
    
    
        
    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    first_multiple_input = input().rstrip().split()

    n = first_multiple_input[0]

    k = int(first_multiple_input[1])

    result = superDigit(n, k)

    fptr.write(str(result) + '\n')

    fptr.close()
```
